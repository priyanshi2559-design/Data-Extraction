1.  Describe different types of data sources used in ETL with suitable examples.

Answer - Data sources are generally categorized into three main types based on their structure -

* Structured data source - These sources have a fixed schema and are highly organized, making them easy to extract and query.
Example - SQL Server, MySQL, Oracle.

* Semi-structured data source - These sources don't have a formal data model but contain tags or markers to seprate data elements. They lack a rigid schema but contain internal structure and metadata.

Example - HTML,XML Files.

* Unstructured data source - These represent the most common type of data but are the most complex for FTL processes because they have no predefined format.
Example - PDF Documents, Images.

CSV - Plain text, only contains raw data

CSV - Faster and more memory-efficient because it lacks complex metadata.
Excel - Slower,requires specialized libraries to parse complex formatting and multiple tabs.

CSV - Preferred for large datasets and high-speed bulk loading into databases.
Excel - Best for human -readable reports or small, manually curated data sources.

CSV - Universal, works across all operating systems and programming languages.
Excel - Proprietary format, tough widely supported, it is more prone to versioning issues.

Summary - In a production ETL pipeline, CSV is usually the  standard because it is lightweight and minimmizes the overhead on the extraction server. Excel is typically only used when the source data is maintained by bussiness users who require formatting and calculations within the file.


4. Explain the steps involved in extracting data from a relational database.

Answer - To extract data from a relational database (like MySQL or PostgreSQL), an ETL developer typically follows these steps:

* Establish Connection: Use a driver (e.g., JDBC/ODBC) and credentials to connect to the source database.

* Identify Change Metadata: Determine if the extraction is a Full Load (all data) or Incremental Load (only new/updated records using timestamps or logs).

* Execute Query: Run SQL SELECT statements to retrieve the required tables and columns.

*  Data Validation: Perform basic checks to ensure the data structure matches the expected schema.

* Stage Data: Move the extracted records into a temporary staging area for the "Transform" phase.

5 .  Explain three common challenges faced during data extraction.

Answer - Common challenges during data extraction -

1. Source System Performance Impact -
Extracting large volumes of data can place a heavy load on production databases. If not managed carefully (e.g., during peak hours), the extraction process can slow down or crash the source system, affecting the end-users of that application.

2. Data Quality and Structural Consistency -
Data sources often contain "dirty data," such as missing values, inconsistent formats (like different date conventions), or duplicate records. Additionally, if the source system undergoes a schema change (e.g., a column is renamed or deleted) without notice, the extraction logic will fail.

3. Connectivity and Security Barriers -
Extracting data from diverse sources—like legacy on-premise systems, third-party APIs, and cloud storage—often involves navigating complex firewall rules, authentication protocols, and varying network speeds. Maintaining a stable, secure connection  across these environments is a constant hurdle for developers.



6. What are APIs? Explain how APIs help in real-time data extraction.

Answer - An API (Application Programming Interface) is a set of rules and protocols that allows different software applications to communicate and exchange data with one another. It acts as an intermediary, enabling one system to request specific information or services from another without needing to know how the underlying system is built.


In the context of real-time data extraction, APIs are essential for the following reasons:

* On-Demand Data Retrieval: Unlike traditional batch processing, which extracts data at scheduled intervals (e.g., daily or weekly), APIs allow systems to request and receive data the moment it is updated or generated.

* Webhooks and Push Notifications: Many APIs support "Webhooks," where the source system automatically "pushes" data to the destination as soon as an event occurs (such as a new sale or a stock price change), ensuring the extraction is instantaneous.

* Reduced Latency: By using standardized protocols like REST or GraphQL, APIs provide a lightweight way to fetch only the necessary data points, minimizing the time it takes to move information from the source to the end-user.

* Automation and Integration: APIs enable developers to automate the flow of data between live services, such as streaming social media feeds or financial market data directly into an analytical dashboard for immediate decision-making.

7. Why are databases preferred for enterprise-level data extraction?

Answer - Databases are preferred for enterprise-level data extraction because they provide a robust infrastructure designed to handle the complexity and scale of corporate environments.

* Efficiency and Performance: Databases utilize indexing and optimized query engines (like SQL), allowing for significantly faster data retrieval compared to scanning large flat files (like CSVs).

* Data Integrity and Relationships: They enforce referential integrity and constraints, ensuring that extracted data remains consistent and accurate across thousands of related tables.

* Scalability and Concurrency: Databases can manage massive volumes of data (terabytes to petabytes) and support hundreds of simultaneous users or processes extracting information without performance degradation.

* Security and Auditing: Enterprise databases offer fine-grained access controls and logging, which are essential for maintaining security and regulatory compliance (like GDPR or HIPAA) during the extraction process.
